# 基础方法+while循环

```python
import cProfile
# 利用基本方法判断一个数是否为质数
def basic_method(n):
    i = 2
    a = True
    if n == 1:
        return False
    while i < n:
        if n % i == 0:
            a = False
            return False
            break
        i += 1
    if a:
        return True

# 性能测试
from timeit import Timer
# __main__代表启动文件
t1 = Timer('basic_method(12345678900987654321)', 'from __main__ import basic_method')
print(t1.timeit())
```

```python
import cProfile
# 利用基本方法判断一个数是否为质数,while-else
def basic_method(n):
    i = 2
    if n == 1:
        return False
    for i in range(2,n):
        if n % i == 0:            
            return False
            break
        i += 1
    else:
        return True

# 性能测试
from timeit import Timer
# __main__代表启动文件
t1 = Timer('basic_method(12345678900987654321)', 'from __main__ import basic_method')
print(t1.timeit())
```

## 失败案例

```python
import cProfile

# 利用除以每个数生成质数列表
def dingyi(end_number, prime_numbers):
    for number in range(1, end_number+1):
        if number % range(1, number) == 0:
            prime_numbers.append(number)

prime_number = []
dingyi(97, prime_number)
```
